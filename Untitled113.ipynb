## Import thư viện

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score
from sklearn import svm
from sklearn.metrics import confusion_matrix
import sklearn
import seaborn as sns


```

```python
train_path = "C:\\Users\\Admin\\Desktop\\DoAnHocMay\\images\\afhq\\train"
val_path = "C:\\Users\\Admin\\Desktop\\DoAnHocMay\\images\\afhq\\val"
types = os.listdir(train_path) 
```

```python
types
```

## Lấy ảnh từ phần train

```python
train_images = []
train_labels = []
mappings =  {'cat':0, 'dog':1, 'wild':2}
for t in types:
    images = os.listdir(train_path +'/'+ t)
    for img_path in images:
        img = Image.open(train_path + '/' + t + '/' + img_path).convert('L')
        img = img.resize((20,20), Image.ANTIALIAS)
        img = np.asarray(img)
        train_images.append(img)
        train_labels.append(mappings[t])
```

## Lấy ảnh từ phần test

```python
val_images = []
val_labels = []
mappings =  {'cat':0, 'dog':1, 'wild':2}
for t in types:
    images = os.listdir(val_path +'/'+ t)
    for img_path in images:
        img = Image.open(val_path +'/'+ t + '/' + img_path).convert('L')
        img = img.resize((20,20), Image.ANTIALIAS)
        img = np.asarray(img)
        val_images.append(img)
        val_labels.append(mappings[t])
```

## Chia ảnh vào X_train , y_train , X_test,y_Test

```python
X_train = np.stack(train_images,axis=0)
y_train = train_labels
X_test = np.stack(val_images,axis=0)
y_test = val_labels

X_train
```

## 2. Sử dụng Multinomial Logistic Regression để dự đoán 

```python
logreg =  sklearn.linear_model.LogisticRegression(C=1e5, solver = 'lbfgs',max_iter=10000, multi_class = 'multinomial')
logreg = logreg.fit(X_train, y_train) # trainning
y_pred = logreg.predict(X_test) # test
```

## 3. Sử dụng Multi-Classes Support Vector Machine để dự đoán

```python
model = svm.SVC(kernel='poly', C=1000000).fit(X_train, y_train)
y_pred_svm = model.predict(X_test)
```
